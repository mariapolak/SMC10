```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Data augementation distribution tests

```python
ps_factors = []
tsm_factors = []
for i in range(100000):
    ps_factor = np.random.uniform(-3, 5)
    ps_factors.append(ps_factor)

    tsm_factor =  np.random.uniform(0.75, 1.35)
    tsm_factors.append(tsm_factor)

plt.figure(figsize=(10, 5))
plt.hist(ps_factors, bins=1000, alpha=0.5, label='Pitch_Shift_Factor')
plt.xlabel('Sample Number')
plt.ylabel('Factor Value')
plt.title('Randomly Generated Factors')
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(tsm_factors, bins=1000, alpha=0.5, label='tsm_factors')
plt.xlabel('Sample Number')
plt.ylabel('Factor Value')
plt.title('Randomly Generated Factors')
plt.show()
```
